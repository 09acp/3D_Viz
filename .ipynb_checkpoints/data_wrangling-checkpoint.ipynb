```python
import pandas as pd 
import numpy as np 
import sqlite3
import missingno as msno # https://github.com/ResidentMario/missingno  examples
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# takes a while 
xls = pd.ExcelFile("WRDS_WhartonFundamentalData.xlsx") # .sheet_names(['sheet1'])
df = xls.parse(0) # by index
print(df.head(2))
```

### EDA 1

##### Observations
- most columns have 100% completeness or above 95% 
- 6 categories are near or below 80% completeness
- some rows have gaps accross multipe features  (data not yet indexed by date)
- high nullite correlation between features 
- possible nullity correation for value imputation based on a regression model


```python
""" See distribution of missing values """
msno.matrix(df)
plt.title('Nullity Distribution by Feature')
plt.show()

""" Observe nullity by column """
msno.bar(df)
plt.title('Feature Completeness')
plt.show()
```

```python
# could be used for value imputation based on a regression model

""" Observe pairwise nullity correlation by feature """
msno.heatmap(df)
plt.title('Pairwise Nullity Correlation')
plt.show()

""" Observe multivariate nullity correlation """
msno.dendrogram(df)
plt.title('Hierarchical Correlation')
plt.show()
```

```python
# Some companies may drop-out from dataset 
print('The number of tickers / companies in the dataset: %s, %s.' %(df['tic'].unique().shape[0], df['conm'].unique().shape[0]))
```

### Wrangle Data

```python
df.columns
"""
Drop Extra Features 
- no documentation for dates 
- these date fields are not valuable for the purposes of this demonstration 
- project will concentrate on the datadate and datafqtr features instead

"""
df.drop( ['rdq','tradedate','fyearq','fqtr','conm','datacqtr'] , axis=1, inplace=True )
# Rename features
df.rename(columns={'tic':'Ticker', 'datafqtr' : 'YQuarter'},inplace=True)
# Create Date features datetime
dates = df['datadate']
df['Date'] = pd.to_datetime(dates, 
               format = '%Y%m%d',
               errors='ignore',
               )
df.drop('datadate',axis = 1, inplace=True)
```
